```python
#load the datasets of the directory 
df_2009 = pd.read_csv('rail-riddle-dataset//2009-01-01 to 2009-12-31  TPT THT Report.csv')
df_2010 = pd.read_csv('rail-riddle-dataset//2010-01-01 to 2010-12-31  TPT THT Report.csv')
df_2011 = pd.read_csv('rail-riddle-dataset//2011-01-01 to 2011-12-31  TPT THT Report.csv')
df_2012 = pd.read_csv('rail-riddle-dataset//2012-01-01 to 2012-12-31  TPT THT Report.csv')
df_2013 = pd.read_csv('rail-riddle-dataset//2013-01-01 to 2013-12-31  TPT THT Report.csv')
df_2014 = pd.read_csv('rail-riddle-dataset//2014-01-01 to 2014-12-31  TPT THT Report.csv')
df_2015 = pd.read_csv('rail-riddle-dataset//2015-01-01 to 2015-12-31  TPT THT Report.csv')
df_2016 = pd.read_csv('rail-riddle-dataset//2016-01-01 to 2016-12-31  TPT THT Report.csv')
df_2017 = pd.read_csv('rail-riddle-dataset//2017-01-01 to 2017-12-31  TPT THT Report.csv')
df_2018 = pd.read_csv('rail-riddle-dataset//2018-01-01 to 2018-12-31  TPT THT Report.csv')
df_2019 = pd.read_csv('rail-riddle-dataset//2019-01-01 to 2019-12-31  TPT THT Report.csv')
df_2020 = pd.read_csv('rail-riddle-dataset//2020-01-01 to 2020-12-31  TPT THT Report.csv')
df_2021 = pd.read_csv('rail-riddle-dataset//2021-01-01 to 2021-12-31  TPT THT Report.csv')
df_dump = pd.read_csv('rail-riddle-dataset\QMASTOR rail data dump.csv')
```

```python

#concatenate all the datasets into one
df = pd.concat([df_2009, df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021], ignore_index=True)
```

```python
#calculate missing values for each column
missing_values_count = df.isnull().sum()
missing_values_count
```

```python
# Calculate the threshold for dropping columns
threshold = 0.5 * df.shape[0]  # 50% of the number of rows
threshold
```

```python
# Drop columns with missing values exceeding the threshold
df = df.dropna(axis=1, thresh=threshold)
```

```python
columns_to_drop = ['Unnamed: 0', 'Plan Note External', 'Actual Notes', 'Rail_ConsignmentNo','T1','SL','FL','T2']
df = df.drop(columns=columns_to_drop)
```

```python
df.columns
```

```python
#impute the missing values with mode for object type and mean for numerical type
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode().iloc[0])
    else:
        df[col] = df[col].fillna(df[col].mean())
```

```python
# Summary statistics for numerical columns
df.describe()
```

```python
#frequency for categorical variables
for col in df.columns:
    if df[col].dtype == 'object':
        print('\n')
        print('Column Name:', col,)
        print(df[col].value_counts())
```

```python
# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Variables for time series analysis
time_series_variables = ['TFR Delay subtracted', 'CNWB_Volume', 'Act_Volume']
```

```python

# Plot time series for relevant variables
plt.figure(figsize=(12, 6))
for variable in time_series_variables:
    plt.plot(df.index, df[variable], label=variable)
plt.title('Time Series of Relevant Variables')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
window = 30  # Adjust window size as needed
for variable in time_series_variables:
    rolling_mean = df[variable].rolling(window=window).mean()
    detrended = df[variable] - rolling_mean
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, detrended, label='Detrended ' + variable)
    plt.title('Detrended Time Series of ' + variable)
    plt.xlabel('Date')
    plt.ylabel('Detrended Value')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#
num_cols = []
cat_cols = []
for col in df.columns:
    if df[col].dtype != 'object':
        num_cols.append(col)
    else:
        cat_cols.append(col)
```

```python
#correlation heatmap of numeric features
corr = df[num_cols].corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1, annot=True, annot_kws={"size": 8}, square=True);
```

```python
# Create bar plots or pie charts for categorical variables
for variable in cat_cols:
    plt.figure(figsize=(8, 6))
    if len(df[variable].unique()) <= 7:
        df[variable].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
        plt.title(f'{variable} Distribution (Pie Chart)')
    elif len(df[variable].unique()) > 7 and len(df[variable].unique()) <= 20:
        sns.countplot(data=df, x=variable)
        plt.title(f'{variable} Distribution (Histogram)')
    else:
        continue
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
from scipy.stats import chi2_contingency

# Analyze the relationship between categorical variables and delays
for variable in cat_cols:
    contingency_table = pd.crosstab(df[variable], df['TFR Delay subtracted'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"Contingency Table for {variable} vs. TFR Delay:")
    print(contingency_table)
    print(f"Chi-Squared Test p-value: {p}\n")
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# Create lag features for delays
num_lags = 10  # Number of lag features
delay_variable = 'TFR Delay subtracted'

for lag in range(1, num_lags+1):
    df[f'{delay_variable}_lag_{lag}'] = df[delay_variable].shift(lag)

# Plot autocorrelation for delay variable
plt.figure(figsize=(10, 6))
plot_acf(df[delay_variable], lags=num_lags)
plt.title(f'Autocorrelation Plot for {delay_variable}')
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.tight_layout()
plt.show()
```
